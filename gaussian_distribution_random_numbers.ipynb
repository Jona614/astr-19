# Importing matplotlib.pylot module and numpy library

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Generates a normal distribution of 1000 random values with a mean of 0 (most values will cluster around the 0) and a standard deviation of 1 (most values will spread and lie close to 0)

# Plot the random_numbers in a histogram; there will be 100 evenly spaced bins between the range 0 to 1

# Name the x and y labels

# Save the figure as a pdf file

```python
def main():
    random_numbers = np.random.normal(0, 1, 1000)
    plt.hist(random_numbers, bins=100)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.savefig('session_10.pdf')

# Calls the main function if the program that is being runned is this exact program
if __name__ == '__main__':
    main()
```
