# Importing matplotlib.pyplot method and numpy library

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Generates a uniform distribution (every value in the specified range has an equal probability of being selected) of 1000 random numbers between 0 to 1

# Plot the random_numbers in a histogram; there will be 100 evenly spaced bins between the range 0 to 1

# Name the x and y labels

# Save the figure as a pdf file

```python
def main():
    random_numbers = np.random.uniform(0, 1, 1000)

    plt.hist(random_numbers, bins=100)

    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.savefig('session_9.pdf')

# Calls the main function if the program that is being runned is this exact program
if __name__ == '__main__':
    main()
```
